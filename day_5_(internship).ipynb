```python
import pandas as pd
df={"studyhours":[2,3,4,5,6,7],"Marks":[40,50,60,70,80,90]}
df=pd.DataFrame(df)
print(df)
```

```python
x=df[["studyhours"]]
y=df[["Marks"]]
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
prediction=model.predict([[8]])
print(prediction)
print(model)
```

```python
df=pd.read_csv('/content/Student.csv')
```

```python
df=pd.DataFrame(df)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
```

```python
df['Attendance']=df['Attendance'].fillna(df['Attendance'].mean())
```

```python
print (df)
```

```python
df=df.drop('Name',axis=1)
print(df)
```

```python
from sklearn.linear_model import LinearRegression
x=df[['Attendance']]
y=df[['Marks']]
model=LinearRegression()
model.fit(x,y)
prediction=model.predict([[92]])
print(prediction)
print(model)
```
